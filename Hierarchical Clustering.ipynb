```python
import pandas as slt
import numpy as prn
import matplotlib.pyplot as tpl
catcher = slt.read_csv("/content/CLUSTERING.csv")
print(catcher.describe())
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(catcher,method = "ward"))
tpl.show()

catcher.describe()
```

```python
hc = AgglomerativeClustering(n_clusters = 5,linkage = "ward")
x =  catcher.iloc[:,[1,2]].values
predict_hc = hc.fit_predict(x)
print(predict_hc)
```

```python
tpl.scatter(x[predict_hc==0,0],x[predict_hc==0,1],c = 'red',label = "Cluster0")
tpl.scatter(x[predict_hc==1,0],x[predict_hc==1,1],c = 'blue',label = "Cluster1")
tpl.scatter(x[predict_hc==2,0],x[predict_hc==2,1],c = 'green',label = "Cluster2")
tpl.scatter(x[predict_hc==3,0],x[predict_hc==3,1],c = 'yellow',label = "Cluster3")
tpl.scatter(x[predict_hc==4,0],x[predict_hc==4,1],c = 'cyan',label = "Cluster4")
tpl.title("Hierarchial Clustering")
```
